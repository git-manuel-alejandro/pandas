# Qué club tiene los mejores jugadores

```python
barcelona = df[df['club'] == "FC Barcelona"]
real = df[df['club'] == "Real Madrid"]
```

```python
real.head()
```

```python
plt.figure()

plt.subplot(121)
plt.boxplot(real["overall"])
plt.ylabel("Overall")
plt.title("Real Madrid")
plt.grid(True)

plt.subplot(122)
plt.boxplot(barcelona["overall"])
plt.title("Barcelona")
plt.grid(True)
```

```python
print('---Real Madrid-----')
print(real["overall"].describe())
print('---Real Madrid-----')
print(barcelona["overall"].describe())
```
