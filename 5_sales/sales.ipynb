```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

#### Merging 12 month of sales data into a single file

```python
df = pd.read_csv('./data/Sales_April_2019.csv')

files = [file for file in os.listdir('./data')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./data/'+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv('all_data.csv', index= False)
    
```

### Read in updated dateframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.shape
```

```python
all_data.head()
```

### Clean up data!

#### Drop NA

```python
all_data = all_data.dropna(how='all')
# nan_df = all_data[all_data.isna().any(axis=1)] count nan
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
# ValueError: invalid literal for int() with base 10: 'Or'
```

#### Convert column to correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with additional columns

### Task 2: Add month column

```python
all_data['month'] = all_data['Order Date'].str[0:2]
all_data['month'] = all_data['month'].astype('int32')
all_data.head()
```

### Add sales columns

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Question 1: What was the best month for sales ? How much was earned that month?

```python
results = all_data.groupby('month').sum('Sales')
results
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.title("2019 Sales per month")
plt.ylabel("Millions")
plt.xlabel("Months")
plt.xticks(months)
plt.show()
```

### Question 2: What city had the highest number of sales?

```python
all_data.head()
```

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

```python
results = all_data.groupby('City').sum('Sales')
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation=70, size= 8)
plt.ylabel("Millions")
plt.xlabel('City Name')
plt.show()
```

#### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data.head()
```

```python
results = all_data.groupby('Hour').sum('Sales').sort_values('Sales', ascending = False)
results
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.title('Sales per hour')
plt.xlabel('Hour')
plt.ylabel('Amount')
plt.grid()
plt.show
```
