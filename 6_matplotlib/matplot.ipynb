```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X, Y)
fig.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x,y , label= '2x')
plt.title('title', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('x label')
plt.ylabel('y label')
plt.xticks(x)
plt.yticks(y)


plt.legend()

plt.savefig('graph.png', dpi=300)
plt.show()
```
